```python
import os
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# === 設定要處理的資料夾 ===
folder_path = "tariff_data"  # ← 改為你的資料夾名稱

# === 初始化情緒分析器 ===
analyzer = SentimentIntensityAnalyzer()

# === 定義情緒分類函式 ===
def get_sentiment(text):
    score = analyzer.polarity_scores(str(text))
    if score['compound'] >= 0.05:
        return 'positive'
    elif score['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# === 處理資料夾內所有 CSV 檔案 ===
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)
        print(f"🔍 處理中：{filename}")

        try:
            df = pd.read_csv(file_path)
            df["sentiment"] = df["Tweet Content"].apply(get_sentiment)

            # 覆寫原始檔案或另存一份（自行選擇）
            output_path = os.path.join(folder_path, filename)  # 覆寫原始
            # output_path = os.path.join(folder_path, f"labeled_{filename}")  # ← 如果你想另存一份加前綴

            df.to_csv(output_path, index=False, encoding='utf-8')
            print(f"✅ 完成標註：{filename}")
        except Exception as e:
            print(f"❌ 錯誤於 {filename}：{e}")
```
