```python
import os
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# === è¨­å®šè¦è™•ç†çš„è³‡æ–™å¤¾ ===
folder_path = "tariff_data"  # â† æ”¹ç‚ºä½ çš„è³‡æ–™å¤¾åç¨±

# === åˆå§‹åŒ–æƒ…ç·’åˆ†æžå™¨ ===
analyzer = SentimentIntensityAnalyzer()

# === å®šç¾©æƒ…ç·’åˆ†é¡žå‡½å¼ ===
def get_sentiment(text):
    score = analyzer.polarity_scores(str(text))
    if score['compound'] >= 0.05:
        return 'positive'
    elif score['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# === è™•ç†è³‡æ–™å¤¾å…§æ‰€æœ‰ CSV æª”æ¡ˆ ===
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)
        print(f"ðŸ” è™•ç†ä¸­ï¼š{filename}")

        try:
            df = pd.read_csv(file_path)
            df["sentiment"] = df["Tweet Content"].apply(get_sentiment)

            # è¦†å¯«åŽŸå§‹æª”æ¡ˆæˆ–å¦å­˜ä¸€ä»½ï¼ˆè‡ªè¡Œé¸æ“‡ï¼‰
            output_path = os.path.join(folder_path, filename)  # è¦†å¯«åŽŸå§‹
            # output_path = os.path.join(folder_path, f"labeled_{filename}")  # â† å¦‚æžœä½ æƒ³å¦å­˜ä¸€ä»½åŠ å‰ç¶´

            df.to_csv(output_path, index=False, encoding='utf-8')
            print(f"âœ… å®Œæˆæ¨™è¨»ï¼š{filename}")
        except Exception as e:
            print(f"âŒ éŒ¯èª¤æ–¼ {filename}ï¼š{e}")
```
